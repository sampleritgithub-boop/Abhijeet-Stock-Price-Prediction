```python


plt.figure(figsize=(12,5))
plt.plot(data["Date"], data["Close"])
plt.title("Tata Steel Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(data["Date"], data["Open"], label="Open")
plt.plot(data["Date"], data["High"], label="High")
plt.plot(data["Date"], data["Low"], label="Low")
plt.plot(data["Date"], data["Close"], label="Close")
plt.legend()
plt.title("OHLC Price Movement")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.bar(data["Date"], data["Volume"])
plt.title("Trading Volume Over Time")
plt.ylabel("Volume")
plt.show()
```

```python
data["Daily_Return"] = data["Close"].pct_change()

data["Daily_Return"].hist(bins=50)
plt.title("Distribution of Daily Returns")
plt.show()
```

```python
import seaborn as sns

corr = data[["Open","High","Low","Close","Volume"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
data["Year"] = data["Date"].dt.year

yearly_avg = data.groupby("Year")["Close"].mean()

yearly_avg.plot(kind="bar", figsize=(10,4))
plt.title("Average Closing Price by Year")
plt.show()
```
